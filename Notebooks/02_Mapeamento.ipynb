### **Variáveis e Loops em Bash**

Variáveis são usadas para armazenar valores como textos ou números e são acessadas com o símbolo $, permitindo reutilizar esses valores ao longo do script.
Já os loops são estruturas que repetem comandos automaticamente, geralmente para processar vários arquivos ou listas de itens. No exemplo, a variável `filename` guarda o nome de um arquivo, e o loop `while read` percorre o arquivo linha por linha, executando comandos repetidamente até o fim do conteúdo. A cada iteração, uma variável contador `(count)` é incrementada, permitindo ao final obter o número total de linhas do arquivo.




````python
%%bash
# Define a variable
filename="GCA_011764305.2_ASM1176430v1.1_assembly_report.txt"
echo "My file is $filename"

# Count lines using a loop
count=0
while read line; do
    count=$((count + 1))
done < "$filename"

echo "Total lines: $count"
````

### **Loops "For" e Wildcards**

Em Bash, wildcards (ou curingas) são símbolos especiais usados para representar padrões de nomes de arquivos, permitindo selecionar múltiplos arquivos de forma flexível. O mais comum é o `*`, que significa “qualquer sequência de caracteres”, então `*.txt` representa todos os arquivos que terminam com extensão `.txt`.  O loop `for` é usado para iterar sobre uma lista de itens — nesse caso, a lista gerada pelo wildcard. Assim, no exemplo, `for file in *.txt` faz com que a variável `file` receba, um por um, os nomes de todos os arquivos `.txt` no diretório. Em cada iteração, o script executa comandos sobre o arquivo atual, como imprimir seu nome e contar o número de linhas com `wc -l`, automatizando o processamento de múltiplos arquivos de forma eficiente.

````python
%%bash
# Create dummy files
touch sample1.txt sample2.txt sample3.txt

# Loop through all .txt files
for file in *.txt
do
    echo "Found file: $file"
    wc -l $file
done
````

## **Seção 4 | Usando programas no Google Colab**


O Google Colab permite a instalação e uso de programas externos através de gerenciadores de pacotes como Conda e pip. No começo de todos os tutorias, será necessário dar acesso ao Google Drive para instalar o Conda e programas necessários para execução dos tutoriais.

> O Conda é um sistema de gerenciamento de pacotes e ambientes muito utilizado em bioinformática por sua praticidade e flexibilidade. Ele permite instalar e gerenciar diferentes programas e bibliotecas de forma padronizada, evitando conflitos de dependências que são comuns em análises genômicas complexas. Além disso, possibilita a criação de ambientes isolados, nos quais cada projeto pode ter suas próprias versões de softwares e pacotes, garantindo maior reprodutibilidade e controle das análises. Em bioinformática, isso é especialmente útil para lidar com ferramentas diversas — como bcftools, samtools, GATK, scikit-allel, Biopython, entre muitas outras — que frequentemente exigem configurações específicas. O Conda, portanto, se tornou uma peça fundamental no fluxo de trabalho de pesquisadores que trabalham com grandes volumes de dados biológicos e precisam de ambientes estáveis e consistentes.

O **Conda** pode ser utilizado no **Google Colab** como uma forma de ampliar e personalizar o ambiente já disponível na nuvem. Embora o Colab venha com diversos pacotes pré-instalados (como NumPy, Pandas e Matplotlib), muitas ferramentas bioinformáticas específicas — por exemplo samtools, bcftools, bedtools, fastqc, minimap2 — não estão disponíveis por padrão. Nesse contexto, o Conda se torna uma solução prática, pois permite instalar rapidamente programas de bioinformática a partir de canais especializados como o bioconda. Além disso, com o Conda é possível criar ambientes isolados dentro da sessão do Colab, garantindo que as versões dos programas utilizados sejam exatamente as necessárias para uma análise reprodutível. Essa flexibilidade é especialmente importante em bioinformática, já que diferentes pipelines podem depender de versões específicas de ferramentas. Assim, o Colab combinado com Conda oferece uma plataforma acessível, gratuita e interativa para realizar análises genômicas, sem a necessidade de infraestrutura local robusta.

## **Seção 6 | Conteúdo adicional: Markdown**

Markdown é uma linguagem de marcação leve e portável para escrever textos formatados de forma simples, legível e versionável. Em bioinformática (e ciência de dados em geral), ele é ideal para documentar análises, criar READMEs de repositórios, explicar códigos e resultados em notebooks (Colab/Jupyter) e gerar relatórios que podem ser facilmente convertidos para HTML/PDF. Por ser texto puro, funciona bem com controle de versão (Git), facilita reprodutibilidade e reduz a dependência de editores pesados — você escreve rápido, mantém tudo junto do seu código e compartilha sem fricção.

### Cheat sheet de Markdown

| Recurso | Sintaxe | Exemplo | Resultado |
|---|---|---|---|
| Título H1 | `# Título` | `# Introdução` | # Introdução |
| Título H2–H6 | `##` … `######` | `### Métodos` | ### Métodos |
| Negrito | `**texto**` ou `__texto__` | `**importante**` | **importante** |
| Itálico | `*texto*` ou `_texto_` | `*ênfase*` | *ênfase* |
| Tachado | `~~texto~~` | `~~remover~~` | ~~remover~~ |
| Código inline | `` `código` `` | `` `ls -lh` `` | `ls -lh` |
| Bloco de código | Três crases acima/abaixo (opcional: linguagem) | ```python<br>print("Olá")<br>``` | bloco de código |
| Citação (blockquote) | `> texto` | `> Nota` | > Nota |
| Lista não ordenada | `- item` / `* item` | `- A`<br>`- B` | • A / B |
| Lista ordenada | `1. item` | `1. A`<br>`2. B` | 1. A / 2. B |
| Lista de tarefas | `- [ ] pendente` / `- [x] feito` | `- [x] Ler` | - [x] Ler |
| Link | `[texto](URL "título")` | `[NCBI](https://ncbi.nlm.nih.gov "Banco")` | [NCBI](https://ncbi.nlm.nih.gov "Banco") |
| Imagem | `![alt](URL "título")` | `![logo](https://exemplo/img.png "Logo")` | (renderiza imagem) |
| Quebra de linha | Dois espaços no fim da linha **ou** `<br>` | `linha 1␠␠`<br>`linha 2` | linha 1<br>linha 2 |
| Regra horizontal | `---` ou `***` | `---` | --- |
| Tabela básica | Cabeçalho + `---` | `\| A \| B \|`<br>`\|---\|---\|`<br>`\| 1 \| 2 \|` | tabela |
| Alinhamento em tabela | `:---` (esq), `:---:` (centro), `---:` (dir) | `\|:---:\|---:\|` | colunas central/à direita |
| Rodapé/nota | `Texto[^1]` e depois `[^1]: nota` | `GC alto[^a]`<br><br>`[^a]: > 50%` | nota de rodapé |
| Escapar caracteres | `\* \_ \# \|` | `\*literal\*` | *literal* |
| Math (se suportado) | `$…$` (inline) ou `$$…$$` (bloco) | `$\\pi r^2$` |  $begin:math:text$ \\pi r^2 $end:math:text$ |
